### 1. Энтропия (3 балла)
Вычислите энтропию и индекс Джини случайной величины $X$, если 

* величина $X$ равновероятно принимает $k\ge2$  значений;

* величина $X$ равномерно распределена на отрезке $[0; a]$;

* величина $X$ нормально распределена $\mathcal N (0, \sigma^2)$;

### 1.1 величина $X$ равновероятно принимает $k\ge2$  значений
Пусть величина X принимает $K$ значений. Вероятности попадания в каждый класс равны и вычисляются как: $$p_{k} = \frac {1}{K}$$
$\textbf{Энтропия равна:}$ $$F_H(R_m)=-\sum_{k = 1}^{K} p_{k} \log_2 p_{k} = - \sum_{k = 1}^{K} \frac {1}{K} \log_2 \frac {1}{K} = - \frac {1}{K}* K *\log_2 \frac {1}{K} = - \log_2 \frac {1}{K} = \log_2 {K}$$
$\textbf{Индекс Джини равен:}$
$$
F_G(R_m)=\sum_{k = 1}^{K} p_{k} (1 - p_{k}) = \sum_{k = 1}^{K} \frac {1}{K} (1 - \frac {1}{K}) = \sum_{k = 1}^{K} \frac {K-1}{K^2} = \frac {K-1}{K} = 1 - \frac {1}{K}
$$

Для непрерывной случайной величины $X$ с плотностью вероятности $f(x)$ определяется дифференциальная энтропия:
$$F_H(R_m) = - \int_{-\infty}^{+\infty} f(x) \log_2 { f(x)}dx$$

### 1.2 величина  $X$ равномерно распределена на отрезке $[0; a]$
Плотность вероятности равномерного распределения: $${f(x)} =  \frac {1}{a-0} = \frac{1}{a}$$ 
$\Rightarrow \textbf{Энтропия равна:}$
$$F_H(R_m) = - \int_{-\infty}^{+\infty}  \frac{1}{a}\log_2 { \frac{1}{a}}dx = -\int_{0}^{a}  \frac{1}{a}\log_2 { \frac{1}{a}}dx  =  - a \frac{1}{a}\log_2 { \frac{1}{a}} = -\log_2 \frac{1}{a} = \log_2 a$$

$\textbf{Индекс Джини:}$
$$ F_G(R_m)=\sum_{k = 1}^{K} p_{k} (1 - p_{k}) = \sum_{k = 1}^{K}  1 (1 - 1) = 0$$
Так как индекс Джини показывает степень отклонения линии фактического распределения от линии равномерного распределения, а в случае равномерного распределения линии совпадают, следовательно, индекс равен 0.

###  2. Критерии информативности (2 балла)
На лекциях говорилось, что критерий информативности для набора объектов R вычисляется на основе того, насколько хорошо их целевые переменные предсказываются константой (при оптимальном выборе этой константы):
$$
H(R)=\frac{1}{|R|}\sum_{(x_i,y_i)\in R} L(y_i,c),
$$
где $L(y, c)$ - некоторая функция потерь и
$$
c=argmin_cH(R).
$$
 
Соответственно, чтобы получить вид критерия при конкретной функции потерь, необходимо аналитически найти оптимальное
значение константы и подставить его в формулу для H(R).
Выведите критерии информативности для следующих функций потерь:

* $L(y,c)=\sum_{k=1}^K (c_k-[y=k])^2$,

* $L(y,c)=-\sum_{k=1}^K [y=k] \log c_k$. 

### 2.1
Минимизируем критерий информативности:
$$H(R) = \min_{c\in \mathbb{Y} } \frac{1}{|R|} \sum_{(x_i,y_i) \in \mathbb{R}} L(y_i,c)$$
$$H(R) = \min_{\sum_{k} c_k = 1} \frac{1}{|R|}\sum_{(x_i,y_i) \in \mathbb{R}} \sum_{k=1}^K (c_k-[y_i=k])^2$$
Найдем производную и приравняем ее к нулю:
$$2\sum_{k=1}^K (c_k-[y=k]) = 0$$
$\Rightarrow$ берем вектор $c_*$, состоящий из долей классов p_k
$$c_* = (p_1,p_2,....., p_k)$$
Подставляем вектор c в критерий информативности:
$$H(R) = \frac{1}{|R|} \sum_{k=1}^K (c_k-p_k)^2 $$
$$H(R) = \frac{1}{|R|} \sum_{k=1}^K (c_k^2-2 c_k p_k +p_k^2) = \frac{1}{|R|} \sum_{k=1}^K (c_k(c_k-2p_k) +p_k^2)=\frac{1}{|R|} (\sum_{k=1}^K c_k(c_k-2p_k) + \sum_{k=1}^K p_k^2)= $$
$$=\frac{1}{|R|} (\sum_{k=1}^K p_k + \sum_{k=1}^K -p_k^2) = =\frac{1}{|R|} (\sum_{k=1}^K p_k - p_k^2)= $$
$$ =\frac{1}{|R|} \sum_{k=1}^K p_k(1-p_k)$$

### 2.2

$$H(R) = \min_{\sum_{k} c_k = 1} \frac{1}{|R|}\sum_{(x_i,y_i) \in \mathbb{R}}\sum_{k=1}^K [y=k] \log c_k$$
Составляем Лагранжиан:
$$L(c, \lambda) = - \frac{1}{|R|}\sum_{(x_i,y_i) \in \mathbb{R}}\sum_{k=1}^K [y=k] \log c_k +\lambda \sum_{k=1}^K c_k\to min_{c_k}$$ 
Дифференцируем:
$$\frac{\partial}{\partial c_k}L(c,\lambda) = -\frac{1}{|R|}\sum_{(x_i,y_i) \in \mathbb{R}}[y_i=k] \frac{1} {c_k} +\lambda = -\frac {p_k}{c_k} +\lambda = 0$$
$$c_k = \frac{p_k}{\lambda}$$
Суммируем равенство по $k$, учитывая, что $\sum_{k=1}^k c_k = 
1$:
$$\sum_{k=1}^k c_k = \sum_{k=1}^k \frac{p_k}{\lambda} = 1$$
$\Rightarrow \lambda = 1 \Rightarrow$ миниум достигается при $c_k = p_k$

$\Rightarrow$ критерий информативности можно переписать как:
$$H(R)=-\sum_{k=1}^K p_k \log P_k$$
